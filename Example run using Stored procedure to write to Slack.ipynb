##  Backup process (no backup but waitfor to show delays)

This code is for a demonstration of a complex process,  updating the status as the code progresses.

The code below will create a post to a slack channel, then will post information below the original post firstly an adhoc progress bar,  and secondly a status of where the process is. Then on completion update the original post to show completion by placing a tick.

  

Additional some information about the run is added to the end of the conversation.

```sql
DECLARE @RawMessage VARCHAR(MAX) 
DECLARE @ChannelId VARCHAR(200) = '#data-bristol'
DECLARE @DatabaseName VARCHAR(200)  = 'Adventure2019'
DECLARE @OriginalThread VARCHAR(200) = ''
DECLARE @ProgressLine VARCHAR(200)
DECLARE @InformationLine VARCHAR(200)
DECLARE @Starttime DATETIME = GETDATE()
DECLARE @endTime DATETIME


SET @RawMessage = ':arrow_forward: The Restore process for ' + @DatabaseName
--Post Message to Slack Channel
EXEC slackToChannel @Channel = @ChannelId OUTPUT, @Message = @RawMessage, @ThreadId = @OriginalThread OUTPUT
-- copy OriginalThread to new parameter as this will be overwritten if passed in to procedure.
SET @progressLine = @OriginalThread
--For information ,  ThreadId of top level post,  and the New ChannelId (required for updating)

--artificial wait to slow process.
waitfor delay '00:00:05'

SET @RawMessage = ':red_circle::red_circle::red_circle::red_circle::red_circle:'
--Post using @Original Thread (new posts will now go below OriginalThread.)
EXEC slackToChannel @Channel = @ChannelId OUTPUT, @Message = @RawMessage, @ThreadId = @progressLine OUTPUT

--For information ,  ThreadId of top level post,  and the New ChannelId (required for updating)

-- copy OriginalThread to new parameter as this will be overwritten if passed in to procedure.
SET @InformationLine = @OriginalThread


SET @RawMessage = 'Starting Full Restore of ' + @DatabaseName
--Post using @Original Thread (new posts will now go below OriginalThread.)
EXEC slackToChannel @Channel = @ChannelId OUTPUT, @Message = @RawMessage, @ThreadId = @InformationLine OUTPUT

--For information ,  ThreadId of top level post,  and the New ChannelId (required for updating)

--Now to update posts with new information.

--artificial wait to slow process.
waitfor delay '00:00:05'

SET @RawMessage = ':large_green_Circle::red_circle::red_circle::red_circle::red_circle:'
--Post using @Original Thread (new posts will now go below OriginalThread.)
EXEC slackToChannel @Channel = @ChannelId , @Message = @RawMessage, @ThreadId = @progressLine , @update = 1

--For information ,  ThreadId of top level post,  and the New ChannelId (required for updating)

-- copy OriginalThread to new parameter as this will be overwritten if passed in to procedure.

SET @RawMessage = 'Full Database Restore completed '
--Post using @Original Thread (new posts will now go below OriginalThread.)
EXEC slackToChannel @Channel = @ChannelId, @Message = @RawMessage, @ThreadId = @InformationLine , @update = 1

waitfor delay '00:00:05'

SET @RawMessage = ':large_green_Circle::large_green_Circle::red_circle::red_circle::red_circle:'
--Post using @Original Thread (new posts will now go below OriginalThread.)
EXEC slackToChannel @Channel = @ChannelId , @Message = @RawMessage, @ThreadId = @progressLine , @update = 1

--For information ,  ThreadId of top level post,  and the New ChannelId (required for updating)

-- copy OriginalThread to new parameter as this will be overwritten if passed in to procedure.

SET @RawMessage = 'restoring Transaction Logs Started '
--Post using @Original Thread (new posts will now go below OriginalThread.)
EXEC slackToChannel @Channel = @ChannelId, @Message = @RawMessage, @ThreadId = @InformationLine , @update = 1


waitfor delay '00:00:05'

SET @RawMessage = ':large_green_Circle::large_green_Circle::large_green_Circle::red_circle::red_circle:'
--Post using @Original Thread (new posts will now go below OriginalThread.)
EXEC slackToChannel @Channel = @ChannelId , @Message = @RawMessage, @ThreadId = @progressLine , @update = 1

--For information ,  ThreadId of top level post,  and the New ChannelId (required for updating)

-- copy OriginalThread to new parameter as this will be overwritten if passed in to procedure.

SET @RawMessage = 'restoring Transaction Logs completed '
--Post using @Original Thread (new posts will now go below OriginalThread.)
EXEC slackToChannel @Channel = @ChannelId, @Message = @RawMessage, @ThreadId = @InformationLine , @update = 1


waitfor delay '00:00:05'

SET @RawMessage = ':large_green_Circle::large_green_Circle::large_green_Circle::large_green_Circle::red_circle:'
--Post using @Original Thread (new posts will now go below OriginalThread.)
EXEC slackToChannel @Channel = @ChannelId , @Message = @RawMessage, @ThreadId = @progressLine , @update = 1

--For information ,  ThreadId of top level post,  and the New ChannelId (required for updating)

-- copy OriginalThread to new parameter as this will be overwritten if passed in to procedure.

SET @RawMessage = 'Putting Database into Recovery '
--Post using @Original Thread (new posts will now go below OriginalThread.)
EXEC slackToChannel @Channel = @ChannelId, @Message = @RawMessage, @ThreadId = @InformationLine , @update = 1

waitfor delay '00:00:05'

SET @RawMessage = ':large_green_Circle::large_green_Circle::large_green_Circle::large_green_Circle::large_green_Circle:'
--Post using @Original Thread (new posts will now go below OriginalThread.)
EXEC slackToChannel @Channel = @ChannelId , @Message = @RawMessage, @ThreadId = @progressLine , @update = 1

--For information ,  ThreadId of top level post,  and the New ChannelId (required for updating)

-- copy OriginalThread to new parameter as this will be overwritten if passed in to procedure.

SET @RawMessage = 'CHECKDB Scan completed '
--Post using @Original Thread (new posts will now go below OriginalThread.)
EXEC slackToChannel @Channel = @ChannelId, @Message = @RawMessage, @ThreadId = @InformationLine , @update = 1


--Set the original post to show a positive outcome.
SET @RawMessage = ':white_check_mark: The Restore process for ' + @DatabaseName
--Post using @Original Thread (new posts will now go below OriginalThread.)
EXEC slackToChannel @Channel = @ChannelId, @Message = @RawMessage, @ThreadId = @OriginalThread , @update = 1


SET @endTime = GETDATE()
--Set the original post to show a positive outcome.
SET @RawMessage = 'Process completed:  in ' + CAST(DATEDIFF(ss,@starttime,@endtime) AS VARCHAR(10)) + ' Seconds'
--Post using @Original Thread (new posts will now go below OriginalThread.)
EXEC slackToChannel @Channel = @ChannelId, @Message = @RawMessage, @ThreadId = @OriginalThread 














```
